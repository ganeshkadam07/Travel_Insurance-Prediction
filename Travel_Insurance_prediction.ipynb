```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

df['Employment Type']=le.fit_transform(df['Employment Type'])
df['GraduateOrNot']=le.fit_transform(df['GraduateOrNot'])
df['FrequentFlyer']=le.fit_transform(df['FrequentFlyer'])
df['EverTravelledAbroad']=le.fit_transform(df['EverTravelledAbroad'])
    
```

```python
df
```

```python
x=df.drop('TravelInsurance',axis=1).values
y=df['TravelInsurance'].values
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=0)

print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler

scalar=StandardScaler()

x_train=scalar.fit_transform(x_train)
x_test=scalar.transform(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

clf=RandomForestClassifier()

param={
    'max_depth':[1,2,3,4,5],
    'max_features':[1,2,3,4,5],
    'max_samples':[1,2,3],
    'n_estimators':[100,150,200,250,300],
    'min_samples_split':[1,2,3],
    'ccp_alpha':[0.0,0.1,0.2,0.3]
}

model=GridSearchCV(clf,param,cv=3)

model.fit(x_train,y_train)
```

```python
model.best_params_
```

```python
model=RandomForestClassifier(ccp_alpha=0.0,
                             max_depth=5,
                             max_features=5,
                             max_samples=3,
                             min_samples_split=2,
                             n_estimators=200
                            )

model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix

print("Acuuracy Score:",accuracy_score(y_test,y_pred))
clf=classification_report(y_test,y_pred)
print("Classification Report:",clf)
```

```python
cm=confusion_matrix(y_test,y_pred)
sns.heatmap(cm,annot=True)
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model=keras.Sequential([
    keras.layers.Dense(30,activation='relu'),
    keras.layers.Dense(10,activation='relu'),
    keras.layers.Dense(1,activation='sigmoid')
])

model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
res=model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=100,batch_size=32)
```

```python
model.evaluate(x_test,y_test)
```
